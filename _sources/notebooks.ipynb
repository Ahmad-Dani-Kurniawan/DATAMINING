```python
# import library
import numpy as np
import pandas as pd
import seaborn as sb
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
#menampilkan data
data = pd.read_csv("https://raw.githubusercontent.com/Ahmad-Dani-Kurniawan/datamining/main/data-jumlah-penduduk-berdasarkan-pendidikan-tahun-2013.csv")
print(data)
```

```python
#menampilkan 5 baris data pertama
data.head()
```

```python
# 1. 
def dt1(i,j):
  P = 6
  Q = 0
  for col in [""nama_provinsi", "nama_kabupaten_kota", "nama_kecamatan", "nama_kelurahan", "pendidikan"]:
    if data[col][i] == data[col][j]:
      Q += 1
  return (P-Q)/P
```

```python
print(dt1(1,2))
print(dt1(1,3))
print(dt1(1,4))
```

```python
def dt2(i,j,h):
  result = 0
  for col in ["jumlah"]:
    result = math.pow(abs(data[col][i] - data[col][j]), h)
  return result**(1/h)
```

```python
# Manhattan distance
print(dt2(1,2,1))
print(dt2(1,3,1))
print(dt2(1,4,1))
```

```python
# Euclidean distance
print(dt2(1,2,2))
print(dt2(1,3,2))
print(dt2(1,4,2))
```
